```python
import pymeshlab as ml

def make_mesh_watertight(input_path, output_path, depth=8):
    """
    Loads a mesh, applies Screened Poisson Reconstruction to make it watertight,
    and saves the cleaned mesh.
    
    :param input_path: Path to the input mesh (obj, ply, stl, etc.)
    :param output_path: Path where the watertight mesh will be saved
    :param depth: Reconstruction depth (higher = more detail, 8-10 is typical)
    """
    # 1. Initialize the MeshSet
    ms = ml.MeshSet()
    
    # 2. Load your messy mesh
    print(f"Loading mesh from: {input_path}")
    ms.load_new_mesh(input_path)
    
    # 3. Compute normals (Poisson reconstruction requires clean vertex normals)
    print("Computing vertex normals...")
    ms.apply_filter('compute_normal_per_vertex')
    
    # 4. Apply Screened Poisson Surface Reconstruction
    # This wraps the mesh and completely seals all holes/intersections.
    print(f"Running Screened Poisson Reconstruction (depth={depth})...")
    ms.apply_filter(
        'generate_surface_reconstruction_screened_poisson',
        depth=depth,
        preclean=True
    )
    
    # Note: Poisson creates a new mesh layer inside the MeshSet. 
    # We want to save this newly generated layer.
    
    # 5. Save the watertight mesh
    print(f"Saving watertight mesh to: {output_path}")
    ms.save_current_mesh(output_path)
    print("Process complete!")

# Example usage:
make_mesh_watertight("../debug_points/room_2_aligned.ply", "../debug_points/room_2_watertight.ply", depth=9)
```
